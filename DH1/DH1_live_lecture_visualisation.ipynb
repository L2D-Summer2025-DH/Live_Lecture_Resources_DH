## DH1: Part 2 - Visualising your data with matplotlib:

```python
from matplotlib.pyplot import subplots

import matplotlib.pyplot as plt
```

```python
fig, ax = subplots()

fig.set_facecolor('lightblue')
ax.set_facecolor('white')
```

### Figure and axes:

```python
from pandas import read_csv

df = read_csv("/Users/Adam/Downloads/patients.csv")

df.head()
```

### Boxplots:

```python
fig, ax = subplots(ncols=3, nrows = 1, figsize=(12, 4))

ax[0].boxplot(df['Age'])
ax[0].set_title('Distribution of Patient Ages')
ax[0].set_ylabel('Age (years)')

ax[1].boxplot(df['Height'])
ax[1].set_title('Distribution of Patient Heights')
ax[1].set_ylabel('Height (inches)')

ax[2].boxplot(df['Weight'])
ax[2].set_title('Distribution of Patient Weights')
ax[2].set_ylabel('Weight (pounds)');

fig.tight_layout()
```

### Histograms:

```python
fig, ax = subplots()

ax.hist(df['Age'], bins=10, edgecolor='black', alpha=0.7)
ax.set_title('Age Distribution Histogram')
ax.set_ylabel('Age (years)');
```

### Scatter Plots:

```python
fig, ax = subplots(figsize=(12, 12))

scatter = ax.scatter(df['Height'], df['Weight'],
                     c=df['Age'], cmap='magma', alpha=1, edgecolor='black', linewidth=0.5)
ax.set_title('Relationship between Height and Weight')
ax.set_ylabel('Weight (pounds)')
ax.set_xlabel('Height (inches)');

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Age (years)')

ax.grid(True, alpha = 0.4)
```
